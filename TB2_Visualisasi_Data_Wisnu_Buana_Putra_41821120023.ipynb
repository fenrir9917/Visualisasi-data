#3. Konsep Kaidah EDA dan filosofinya

Exploratory Data Analysis (EDA) adalah pendekatan untuk menganalisis data yang melibatkan penggunaan teknik statistik, visualisasi, dan analisis data lainnya untuk memahami karakteristik data secara mendalam.

Prinsip filosofi EDA adalah bagaimana kita membedah data.

1.   Apa yang ingin kita ketahui
2.   Bagaimana kita melihat data
3.   Bagaimana kita menafsirkan data

EDA bertujuan untuk mendapatkan wawasan yang kaya dan pemahaman yang lebih baik tentang data yang sedang dianalisis, agar dapat membuat keputusan yang lebih tepat dan menggali pengetahuan baru dari data yang ada.

#4. Penjelasan Konsep Univariate, bivariate, multivariate

a. Univariate: Analisis univariate adalah analisis yang melibatkan satu variabel tunggal dalam sebuah dataset. Tujuannya adalah untuk memahami karakteristik variabel tersebut tanpa mempertimbangkan hubungannya dengan variabel lain.

b. Bivariate: Analisis bivariate melibatkan dua variabel dalam sebuah dataset. Fokusnya adalah pada hubungan atau korelasi antara kedua variabel tersebut.

c. Multivariate: Analisis multivariate melibatkan lebih dari dua variabel dalam sebuah dataset. Tujuannya adalah untuk memahami hubungan dan interaksi antara variabel-variabel tersebut secara simultan.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('/content/drive/MyDrive/Visualisasi_data/dataset/employee.csv')
df.head()
```

```python
df.drop(['EmployeeNumber', 'Over18', 'StandardHours'], axis="columns", inplace=True)
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
df["Attrition"] = label.fit_transform(df.Attrition)
```

```python
df.info()
```

```python
#confusion matrix
plt.figure(figsize=(15, 15))
sns.heatmap(df.corr(), annot=True, cmap="RdYlGn", annot_kws={"size":7})
```

```python
# Menghitung matriks korelasi
corr_matrix = df.corr()

# Mengurutkan matriks korelasi secara ascending
corr_sorted = corr_matrix.iloc[0].sort_values()

# Visualisasi korelasi dengan diagram batang
fig, ax = plt.subplots(figsize=(10, 8))
bars = ax.barh(corr_sorted.index, corr_sorted.values, align='center')

# Menambahkan label nilai ke dalam diagram batang
for bar in bars:
    value = bar.get_width()
    ax.text(value, bar.get_y() + bar.get_height() / 2, f'{value:.2f}', ha='left', va='center')

ax.set_xlabel('Korelasi')
ax.set_ylabel('Fitur')
ax.set_title('Visualisasi Korelasi')
plt.show()
```

```python
#prepare the training set
X = df.iloc[:, :-1]
Y = df.iloc[:, -1]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.model_selection import GridSearchCV

X = df.drop('Attrition', axis=1)
y = df['Attrition']

# Perform one-hot encoding for categorical variables
categorical_cols = X.select_dtypes(include='object').columns
encoder = OneHotEncoder(sparse_output=False, drop='first')
X_encoded = pd.DataFrame(encoder.fit_transform(X[categorical_cols]))
X_encoded.columns = encoder.get_feature_names_out(categorical_cols)
X.drop(categorical_cols, axis=1, inplace=True)
X = pd.concat([X, X_encoded], axis=1)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Train the Logistic Regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2%}")
print(classification_report(y_test, y_pred))
```

```python
y_test.value_counts()[0]/ y_test.shape[0]
```

```python
from sklearn.metrics import precision_recall_curve, roc_curve, roc_auc_score

# Mendapatkan probabilitas prediksi kelas positif
y_prob = model.predict_proba(X_test)[:, 1]

# Menghitung precision, recall, dan threshold
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

# Plot precision-recall vs threshold
plt.figure(figsize=(8, 6))
plt.plot(thresholds, precision[:-1], 'b--', label='Precision')
plt.plot(thresholds, recall[:-1], 'g--', label='Recall')
plt.xlabel('Threshold')
plt.ylabel('Precision-Recall')
plt.title('Precision-Recall vs Threshold')
plt.legend()
plt.grid(True)
plt.show()

# Menghitung false positive rate, true positive rate, dan threshold
fpr, tpr, thresholds_roc = roc_curve(y_test, y_prob)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, 'b-', label='ROC Curve')
plt.plot([0, 1], [0, 1], 'r--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.grid(True)
plt.show()

# Menghitung nilai AUC-ROC
roc_auc = roc_auc_score(y_test, y_prob)
print(f"ROC AUC Score: {roc_auc:.4f}")
```
